## Intorduction:

This is a synthetic dataset about women with Polycystic Ovary Syndrome (PCOS), generated according to the Rotterdam Criteria. The Rotterdam Criteria, established in 2003 by an international consensus of experts, defines that, for a woman to be diagnosed with the syndrome, she must meet at least two of the three criteria.

Since this is a synthetic dataset, some peculiarities may be present. Therefore, it is essential that the analysis be accompanied by a medical specialist, especially for commercial use.

The categorical numeric columns (Irregular Menstruation and PCOS Diagnosis) are coded in binary format, where 0 represents "No" and 1 represents "Yes".

## Purpose:

My goal in conducting exploratory data analysis (EDA) on this dataset is to develop a machine learning model capable of predicting Polycystic Ovary Syndrome (PCOS) based on new data points. By thoroughly analyzing the dataset, I aim to uncover patterns and insights that will enhance the model's accuracy and reliability. This foundational work will be essential for making informed predictions in future cases.

## Columns Details:

-   Age: The age of the woman in years. This is an important demographic factor that can influence health conditions.
-   BMI (Body Mass Index): A measure calculated from a person's height and weight.
-   Menstruation Irregularity: Indicates whether the woman experiences irregular menstrual cycles. 0 represents "Irregularity" and 1 represents "Regularity".
-   Testtrone Level (ng/dl): The concentration of testosterone in the blood, measured in nanograms per deciliter.
-   Antral Follicle Count: A measure of the number of antral follicles (small fluid-filled sacs in the ovaries) visible during an ultrasound./
-   PCOS-Diagnosis: 0 represent "No", and 1 represents "Yes".
